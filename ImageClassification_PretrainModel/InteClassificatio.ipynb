```python
import numpy as np
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt
import glob as gb
import tensorflow as tf
from tensorflow.keras.layers import Input, Lambda, Dense, Flatten, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.applications.vgg19 import VGG19
from tensorflow.keras.applications.vgg19 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential, Input, Model
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.utils import load_img
from keras.layers.advanced_activations import LeakyReLU
```

```python
train_path = "archive/seg_train/seg_train/"
test_path = "archive/seg_test/seg_test/"
pred_path = "archive/seg_pred/seg_pred/"
```

```python
#####Training data####
for folder in os.listdir(train_path):
    files = gb.glob(pathname=str(train_path + folder + "/*.jpg"))
    print(f"{len(files)} in folder {folder}")  # print length for each class
```

```python
#####Testing data####
for folder in os.listdir(test_path):
    files = gb.glob(pathname=str(test_path + folder + "/*.jpg"))
    print(f"{len(files)} in folder {folder}")  # print length for each class
```

```python
count = 0
for files in os.listdir(pred_path):
    count = count + 1
print(f"There are {count} files in pred directory")
```

```python
categories = {
    "buildings": 0,
    "forest": 1,
    "glacier": 2,
    "mountain": 3,
    "sea": 4,
    "street": 5,
}

def getcategories(n) : 
    for x , y in categories.items() : 
        if n == y : 
            return x    #return the name of categories
```

```python
x_train = []
y_train = []

for folder in os.listdir(train_path):
    files = gb.glob(pathname=str(train_path + folder + "/*.jpg"))
    for file in files:
        image = cv2.imread(file)
        x_train.append(
            list(cv2.resize(image, (224, 224)))
        )  # list all images after resize in x_train
        y_train.append(categories[folder])  # get the names of images from code function
```

```python
plt.figure(figsize=(15, 15))
for n, i in enumerate(
    list(np.random.randint(0, len(x_train), 9))
):  # print random images
    plt.subplot(3, 3, n + 1)
    plt.imshow(x_train[i])
    plt.axis("off")
    plt.title(getcategories(y_train[i]))
```

```python
x_test = []
y_test = []

for folder in os.listdir(test_path):
    files = gb.glob(pathname=str(test_path + folder + "/*.jpg"))
    for file in files:
        image = cv2.imread(file)
        x_test.append(
            list(cv2.resize(image, (224, 224)))
        )  # list all images after resize in x_train
        y_test.append(categories[folder])  # get the names of images from code function
```

```python
x_pred = []
files = gb.glob(
    pathname=str(pred_path + "/*.jpg")
)  # go on seg_pred path and get all files jpg
for file in files:
    image = cv2.imread(file)
    x_pred.append(
        list(cv2.resize(image, (224, 224)))
    )  # list all images after resize in x_pred
```

```python
plt.figure(figsize=(10, 10))
for n, i in enumerate(list(np.random.randint(0, len(x_pred), 9))):
    plt.subplot(3, 3, n + 1)
    plt.imshow(x_pred[i])
    plt.axis("off")
```

```python
x_train = np.array(x_train)
x_test = np.array(x_test)
x_pred = np.array(x_pred)
y_train = np.array(y_train)
y_test = np.array(y_test)
x_train = x_train.astype("float32")
x_test = x_test.astype("float32")
x_pred = x_pred.astype("float32")
```

```python
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    zoom_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode="nearest",
)
train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=(224, 224),
    batch_size=1,
    class_mode="categorical",
    shuffle=True,
)
```

```python
val_datagen = ImageDataGenerator(rescale=1.0 / 255)
val_generator = val_datagen.flow_from_directory(
    test_path,
    target_size=(224, 224),
    batch_size=1,
    class_mode="categorical",
    shuffle=True,
)
```

```python
vgg19_model = VGG19(weights="imagenet", include_top=False, input_shape=(224, 224, 3))
```

```python

for layer in vgg19_model.layers[:14]:
    layer.trainable = False
```

```python
x = vgg19_model.output
x = Flatten()(x) 
x = Dense(500, activation="relu")(x)
x = Dropout(0.5)(x) 
x = Dense(250, activation="relu")(x)
x = Dropout(0.5)(x)  
x = Dense(6, activation="softmax")(x) 
model = Model(inputs=vgg19_model.input, outputs=x)
```

```python
from keras.callbacks import ReduceLROnPlateau

lr_reduce = ReduceLROnPlateau(
    monitor="val_accuracy", factor=0.6, patience=8, verbose=1, mode="max",
)
```

```python
from tensorflow.keras import layers, models, Model, optimizers

learning_rate = 5e-5
model.compile(
    loss="categorical_crossentropy",
    optimizer=optimizers.Adam(lr=learning_rate),
    metrics=["accuracy"],
)
history = model.fit(
    train_generator,
    batch_size=50,
    epochs=10,
    validation_data=val_generator,
    callbacks=[lr_reduce],
)
```

```python
plt.plot(history.history["loss"], label="training loss")
plt.plot(history.history["val_loss"], label="validation loss")
plt.legend()
plt.show()
```

```python
plt.plot(history.history["accuracy"], label="accuracy")
plt.plot(history.history["val_accuracy"], label="validation accuracy")
plt.legend()
plt.show()
```
